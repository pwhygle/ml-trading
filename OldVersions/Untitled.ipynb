```python
from typing import Any
from dataclasses import dataclass
import json
@dataclass
class CurrentBalances:
    accruedInterest: float
    cashBalance: float
    cashReceipts: float
    longOptionMarketValue: float
    liquidationValue: float
    longMarketValue: float
    moneyMarketFund: float
    savings: float
    shortMarketValue: float
    pendingDeposits: float
    availableFunds: float
    availableFundsNonMarginableTrade: float
    buyingPower: float
    buyingPowerNonMarginableTrade: float
    dayTradingBuyingPower: float
    equity: float
    equityPercentage: float
    longMarginValue: float
    maintenanceCall: float
    maintenanceRequirement: float
    marginBalance: float
    regTCall: float
    shortBalance: float
    shortMarginValue: float
    shortOptionMarketValue: float
    sma: float
    mutualFundValue: float
    bondValue: float

    @staticmethod
    def from_dict(obj: Any) -> 'CurrentBalances':
        _accruedInterest = float(obj.get("accruedInterest"))
        _cashBalance = float(obj.get("cashBalance"))
        _cashReceipts = float(obj.get("cashReceipts"))
        _longOptionMarketValue = float(obj.get("longOptionMarketValue"))
        _liquidationValue = float(obj.get("liquidationValue"))
        _longMarketValue = float(obj.get("longMarketValue"))
        _moneyMarketFund = float(obj.get("moneyMarketFund"))
        _savings = float(obj.get("savings"))
        _shortMarketValue = float(obj.get("shortMarketValue"))
        _pendingDeposits = float(obj.get("pendingDeposits"))
        _availableFunds = float(obj.get("availableFunds"))
        _availableFundsNonMarginableTrade = float(obj.get("availableFundsNonMarginableTrade"))
        _buyingPower = float(obj.get("buyingPower"))
        _buyingPowerNonMarginableTrade = float(obj.get("buyingPowerNonMarginableTrade"))
        _dayTradingBuyingPower = float(obj.get("dayTradingBuyingPower"))
        _equity = float(obj.get("equity"))
        _equityPercentage = float(obj.get("equityPercentage"))
        _longMarginValue = float(obj.get("longMarginValue"))
        _maintenanceCall = float(obj.get("maintenanceCall"))
        _maintenanceRequirement = float(obj.get("maintenanceRequirement"))
        _marginBalance = float(obj.get("marginBalance"))
        _regTCall = float(obj.get("regTCall"))
        _shortBalance = float(obj.get("shortBalance"))
        _shortMarginValue = float(obj.get("shortMarginValue"))
        _shortOptionMarketValue = float(obj.get("shortOptionMarketValue"))
        _sma = float(obj.get("sma"))
        _mutualFundValue = float(obj.get("mutualFundValue"))
        _bondValue = float(obj.get("bondValue"))
        return CurrentBalances(_accruedInterest, _cashBalance, _cashReceipts, _longOptionMarketValue, _liquidationValue, _longMarketValue, _moneyMarketFund, _savings, _shortMarketValue, _pendingDeposits, _availableFunds, _availableFundsNonMarginableTrade, _buyingPower, _buyingPowerNonMarginableTrade, _dayTradingBuyingPower, _equity, _equityPercentage, _longMarginValue, _maintenanceCall, _maintenanceRequirement, _marginBalance, _regTCall, _shortBalance, _shortMarginValue, _shortOptionMarketValue, _sma, _mutualFundValue, _bondValue)

@dataclass
class InitialBalances:
    accruedInterest: float
    availableFundsNonMarginableTrade: float
    bondValue: float
    buyingPower: float
    cashBalance: float
    cashAvailableForTrading: float
    cashReceipts: float
    dayTradingBuyingPower: float
    dayTradingBuyingPowerCall: float
    dayTradingEquityCall: float
    equity: float
    equityPercentage: float
    liquidationValue: float
    longMarginValue: float
    longOptionMarketValue: float
    longStockValue: float
    maintenanceCall: float
    maintenanceRequirement: float
    margin: float
    marginEquity: float
    moneyMarketFund: float
    mutualFundValue: float
    regTCall: float
    shortMarginValue: float
    shortOptionMarketValue: float
    shortStockValue: float
    totalCash: float
    isInCall: bool
    pendingDeposits: float
    marginBalance: float
    shortBalance: float
    accountValue: float

    @staticmethod
    def from_dict(obj: Any) -> 'InitialBalances':
        _accruedInterest = float(obj.get("accruedInterest"))
        _availableFundsNonMarginableTrade = float(obj.get("availableFundsNonMarginableTrade"))
        _bondValue = float(obj.get("bondValue"))
        _buyingPower = float(obj.get("buyingPower"))
        _cashBalance = float(obj.get("cashBalance"))
        _cashAvailableForTrading = float(obj.get("cashAvailableForTrading"))
        _cashReceipts = float(obj.get("cashReceipts"))
        _dayTradingBuyingPower = float(obj.get("dayTradingBuyingPower"))
        _dayTradingBuyingPowerCall = float(obj.get("dayTradingBuyingPowerCall"))
        _dayTradingEquityCall = float(obj.get("dayTradingEquityCall"))
        _equity = float(obj.get("equity"))
        _equityPercentage = float(obj.get("equityPercentage"))
        _liquidationValue = float(obj.get("liquidationValue"))
        _longMarginValue = float(obj.get("longMarginValue"))
        _longOptionMarketValue = float(obj.get("longOptionMarketValue"))
        _longStockValue = float(obj.get("longStockValue"))
        _maintenanceCall = float(obj.get("maintenanceCall"))
        _maintenanceRequirement = float(obj.get("maintenanceRequirement"))
        _margin = float(obj.get("margin"))
        _marginEquity = float(obj.get("marginEquity"))
        _moneyMarketFund = float(obj.get("moneyMarketFund"))
        _mutualFundValue = float(obj.get("mutualFundValue"))
        _regTCall = float(obj.get("regTCall"))
        _shortMarginValue = float(obj.get("shortMarginValue"))
        _shortOptionMarketValue = float(obj.get("shortOptionMarketValue"))
        _shortStockValue = float(obj.get("shortStockValue"))
        _totalCash = float(obj.get("totalCash"))
        _isInCall = bool(obj.get("isInCall"))
        _pendingDeposits = float(obj.get("pendingDeposits"))
        _marginBalance = float(obj.get("marginBalance"))
        _shortBalance = float(obj.get("shortBalance"))
        _accountValue = float(obj.get("accountValue"))
        return InitialBalances(_accruedInterest, _availableFundsNonMarginableTrade, _bondValue, _buyingPower, _cashBalance, _cashAvailableForTrading, _cashReceipts, _dayTradingBuyingPower, _dayTradingBuyingPowerCall, _dayTradingEquityCall, _equity, _equityPercentage, _liquidationValue, _longMarginValue, _longOptionMarketValue, _longStockValue, _maintenanceCall, _maintenanceRequirement, _margin, _marginEquity, _moneyMarketFund, _mutualFundValue, _regTCall, _shortMarginValue, _shortOptionMarketValue, _shortStockValue, _totalCash, _isInCall, _pendingDeposits, _marginBalance, _shortBalance, _accountValue)

@dataclass
class ProjectedBalances:
    availableFunds: float
    availableFundsNonMarginableTrade: float
    buyingPower: float
    dayTradingBuyingPower: float
    dayTradingBuyingPowerCall: float
    maintenanceCall: float
    regTCall: float
    isInCall: bool
    stockBuyingPower: float

    @staticmethod
    def from_dict(obj: Any) -> 'ProjectedBalances':
        _availableFunds = float(obj.get("availableFunds"))
        _availableFundsNonMarginableTrade = float(obj.get("availableFundsNonMarginableTrade"))
        _buyingPower = float(obj.get("buyingPower"))
        _dayTradingBuyingPower = float(obj.get("dayTradingBuyingPower"))
        _dayTradingBuyingPowerCall = float(obj.get("dayTradingBuyingPowerCall"))
        _maintenanceCall = float(obj.get("maintenanceCall"))
        _regTCall = float(obj.get("regTCall"))
        _isInCall = bool(obj.get("isInCall"))
        _stockBuyingPower = float(obj.get("stockBuyingPower"))
        return ProjectedBalances(_availableFunds, _availableFundsNonMarginableTrade, _buyingPower, _dayTradingBuyingPower, _dayTradingBuyingPowerCall, _maintenanceCall, _regTCall, _isInCall, _stockBuyingPower)


@dataclass
class SecuritiesAccount:
    type: str
    accountId: str
    roundTrips: int
    isDayTrader: bool
    isClosingOnlyRestricted: bool
    initialBalances: InitialBalances
    currentBalances: CurrentBalances
    projectedBalances: ProjectedBalances

    @staticmethod
    def from_dict(obj: Any) -> 'SecuritiesAccount':
        _type = str(obj.get("type"))
        _accountId = str(obj.get("accountId"))
        _roundTrips = int(obj.get("roundTrips"))
        _isDayTrader = bool(obj.get("isDayTrader"))
        _isClosingOnlyRestricted =  bool(obj.get("isClosingOnlyRestricted"))
        _initialBalances = InitialBalances.from_dict(obj.get("initialBalances"))
        _currentBalances = CurrentBalances.from_dict(obj.get("currentBalances"))
        _projectedBalances = ProjectedBalances.from_dict(obj.get("projectedBalances"))
        return SecuritiesAccount(_type, _accountId, _roundTrips, _isDayTrader, _isClosingOnlyRestricted, _initialBalances, _currentBalances, _projectedBalances)

# Example Usage
# jsonstring = json.loads(myjsonstring)
# root = Root.from_dict(jsonstring)
#Assume you received this JSON response
@dataclass
class Root:
    securitiesAccount: SecuritiesAccount

    @staticmethod
    def from_dict(obj: Any) -> 'Root':
        _securitiesAccount = SecuritiesAccount.from_dict(obj.get("securitiesAccount"))
        return Root(_securitiesAccount)
```
